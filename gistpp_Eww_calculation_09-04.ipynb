# Calculating Eww using gistpp

```python
import os
import subprocess
from decimal import Decimal


temperatures = [f"{t}" for t in range(260, 361, 20)]
distances = [f"{Decimal('0.24') + Decimal('0.02') * i:.2f}" for i in range(15, 64)]


refdens = {
    "260": 0.03228295317,
    "280": 0.03343042365,
    "300": 0.03331284139,
    "320": 0.03306310045,
    "340": 0.03257995556,
    "360": 0.03228279553
}

input_path = '/Users/arghavan/Graduate Center Dropbox/Arghavan Vedadi Gargari/MyFiles/GIST_data/'
output_path = '/Users/arghavan/lab/hp-results/Eww_data/'

for t in temperatures:
    output_file = f"{output_path}{t}K_Eww.dx"
    dens = refdens[t]

    with open(output_file, 'a') as dx_file:
        for d in distances:

            dir_path = f"{input_path}{t}K/{float(d)*100}/"
            os.chdir(dir_path)
        
            Eww_tot = subprocess.run(f"gistpp -i gZ-Eww-dens.dx -op sum", shell=True, capture_output=True, text=True)
            E_lines = Eww_tot.stdout.splitlines()
            E_sum = float(E_lines[0].split()[-1])

            gO_tot = subprocess.run("gistpp -i gZ-gO.dx -op sum", shell=True, capture_output=True, text=True)
            gO_lines = gO_tot.stdout.splitlines()
            gO_sum = float(gO_lines[0].split()[-1])

            Eww_per_molecule = E_sum/(gO_sum*dens)
            print (f'Eww_per_molecule for {d} at {t}K: {Eww_per_molecule} Kcal/mol')
            dx_file.write(f"{d} {Eww_per_molecule}\n")

```

```python
t = '260'
d = '1.50'


refdens = {
    "260": 0.03228295317,
    "280": 0.03343042365,
    "300": 0.03331284139,
    "320": 0.03306310045,
    "340": 0.03257995556,
    "360": 0.03228279553
}





base_path = '/gibbs/arghavan/gist_0.1/'

dir_path = f"{base_path}GIST_0.1_{t}K/{t}K_{d}/"
out_path = f"{base_path}GIST_0.1_{t}K/"
dens = refdens[t]

os.chdir(dir_path)

Eww_tot = subprocess.run(f"gistpp -i gZ-Eww-dens.dx -op sum", shell=True, capture_output=True, text=True)
gO_tot = subprocess.run("gistpp -i gZ-gO.dx -op sum", shell=True, capture_output=True, text=True)


E_lines = Eww_tot.stdout.splitlines()
E_sum = float(E_lines[0].split()[-1])

gO_lines = gO_tot.stdout.splitlines()
gO_sum = float(gO_lines[0].split()[-1])


Eww_per_molecule = E_sum/(gO_sum*dens)
print (f'Eww_per_molecule for d = {d} at {t}K: {Eww_per_molecule} Kcal/mol')
```

```python

import pickle
E_list = []
E_list.append([d, Eww_per_molecule])

pkl_file = f"{base_path}GIST_0.1_{t}K/{t}K_Eww.pkl"
with open(pkl_file, 'ab') as array_file:
    pickle.dump(E_list, array_file)
```
