```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
```

```python
input_path = '/Users/arghavan/lab/hp-results/meniscus-results/300K/0.60/'

numdens_y_data = np.loadtxt(f"{input_path}numdens-y.dat")
numdens_z_data = np.loadtxt(f"{input_path}numdens-z.dat")


distances_y = numdens_y_data[:, 0]
distances_z = numdens_z_data[:, 0] 
numdens_y= numdens_y_data[:, 1]  # second column
numdens_z= numdens_z_data[:, 1]  # second column


bulk_density = 0.03331284139 
y_ref = numdens_y / bulk_density
z_ref = numdens_z / bulk_density

plt.figure(figsize=(8, 6))
plt.plot(distances_y, y_ref, color='blue', linewidth=2, label=f"Plates distance = 6.0 Å")
# plt.plot(bin_centers, rho_i_plate, color='green', linewidth=2, label=f"{float(d)*10:.1f} Å")
plt.title("Number Density Distribution Along Y-axis")
plt.xlabel('Distance from the edge in Å', fontsize=14)
plt.ylabel('Number density (referenced to bulk)', fontsize=14)
plt.legend()
plt.xlim(0, 18)
plt.ylim(bottom=0)
# plt.xticks([-4.0, -2.0, 0.0, 2.0])
plt.gca().xaxis.set_minor_locator(MultipleLocator(0.5))
plt.tick_params(axis='x', which='minor', length=2)
plt.tight_layout()
# plt.savefig(f"{output_path}/numDens-y-ref.png", dpi=300)
# plt.close()


plt.figure(figsize=(8, 6))
plt.plot(distances_z, z_ref, color='blue', linewidth=2, label=f"Plates distance = 6.0 Å")
plt.title("Number Density Distribution Along Z-axis")
plt.xlabel('Distance from the edge in Å', fontsize=14)
plt.ylabel('Number density (referenced to bulk)', fontsize=14)
plt.legend()
plt.xlim(0, 18)
plt.ylim(bottom=0)
# plt.xticks([-4.0, -2.0, 0.0, 2.0])
plt.gca().xaxis.set_minor_locator(MultipleLocator(0.5))
plt.tick_params(axis='x', which='minor', length=2)
plt.tight_layout()
# plt.savefig(f"{output_path}/numDens-z-ref.png", dpi=300)
# plt.close()
```
